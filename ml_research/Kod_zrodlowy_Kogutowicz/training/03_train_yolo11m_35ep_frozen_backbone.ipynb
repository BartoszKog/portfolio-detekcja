```python
from ultralytics import YOLO
from pathlib import Path
```

# Trening model m

```python
data_yaml = Path("data_1024_res/data.yaml")
```

```python
project_path = 'training_results'
name_run = 'yolo11m_freeze10'

model = YOLO("yolo11m-obb.pt")
results = model.train(
    data=data_yaml,
    epochs=50,
    imgsz=1024,
    batch=-1,
    patience=10,
    single_cls=True,
    project=project_path,
    name=name_run,
    exist_ok=True,
    freeze=10
)
```

Po 30. epoce nastąpił mechanizm EarlyStopping, należy powtórzyć trening na mniejszej liczbie epok aby zakończyć trening bez mozaikowania.

```python
project_path = 'training_results'
name_run = 'yolo11m_freeze10_epochs35'

model = YOLO("yolo11m-obb.pt")
results = model.train(
    data=data_yaml,
    epochs=35,
    imgsz=1024,
    batch=-1,
    patience=10,
    single_cls=True,
    project=project_path,
    name=name_run,
    exist_ok=True,
    freeze=10
)
```
