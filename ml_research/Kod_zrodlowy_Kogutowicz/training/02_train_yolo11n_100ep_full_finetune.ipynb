```python
from ultralytics import YOLO
from pathlib import Path
```

# Trening model n

```python
data_yaml = Path("data_1024_res/data.yaml")
```

```python
project_path = 'training_results'
name_run = 'yolo11n_epchos100'

model = YOLO("yolo11n-obb.pt")
results = model.train(
    data=data_yaml,
    epochs=100,
    imgsz=1024,
    batch=-1,
    patience=20,
    single_cls=True,
    project=project_path,
    name=name_run,
    exist_ok=True,
)
```
